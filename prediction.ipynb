```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pandas as pd
import numpy as np
import pickle
```

```python
## load trained model, scaler, onehot encoder, label encoder
model = load_model('/content/drive/MyDrive/ANN CLASSIFICATION/model.h5')
```

```python
## Load encoder and scaler
with open('/content/drive/MyDrive/ANN CLASSIFICATION/label_encoder_gender.pkl','rb') as file:
  label_encoder_gender = pickle.load(file)

with open('/content/drive/MyDrive/ANN CLASSIFICATION/onehot_encoder_geo.pkl','rb') as file:
  onehot_encoder_geo = pickle.load(file)

with open('/content/drive/MyDrive/ANN CLASSIFICATION/scaler.pkl','rb') as file:
  scaler = pickle.load(file)
```

```python
## Input Data
input_data = {
    'CreditScore': 600,
    'Geography': "France",
    'Gender': "Male",
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
# Convert input data into dataframe

input_df = pd.DataFrame([input_data])
input_df
```

```python
## OneHot Encode 'Geography
geo_encoded = onehot_encoder_geo.transform([[input_data['Geography']]])
geo_encoded_df = pd.DataFrame(geo_encoded.toarray(),columns = onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
## Combine OneHot Encoded value with input_data
input_df = pd.concat([input_df.drop('Geography',axis=1),geo_encoded_df],axis=1)
input_df
```

```python
## label encode categorical variables (gender)
input_df['Gender'] = label_encoder_gender.transform(input_df['Gender'])
input_df
```

```python
## Scaling the input_data
scaled_input = scaler.transform(input_df)
```

```python
scaled_input
```

```python
## Prediction
prediction = model.predict(scaled_input)
prediction
```

```python
prediction_proab = prediction[0][0]
prediction_proab
```

```python
if prediction_proab > 0.5:
  print("The customer is likely to churn.")
else:
  print("The customer is not likely to churn.")
```
